```python
data = df.filter(['Open'])
print(data.head())
print("\n")
dataset = data.values
training_data_len = int(np.ceil(len(dataset) * .8))
print(training_data_len)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(dataset)
scaled_data
```

```python
train_x = []
train_y = []
train_data = scaled_data[0:int(training_data_len),:]

for i in range(60,len(train_data)):
    train_x.append(train_data[i-60:i,0])
    train_y.append(train_data[i,0])
    if i<= 61:
        print(train_x)
        print(train_y)
```

```python
train_x,train_y = np.array(train_x) , np.array(train_y)
```

```python
train_x.shape
```

```python
train_x = np.reshape(train_x,(train_x.shape[0], train_x.shape[1], 1))
train_x.shape
```

```python
model = Sequential()
model.add(LSTM(units = 50 , return_sequences=True , input_shape = (train_x.shape[1],1)))
model.add(LSTM(units = 50 , return_sequences = True))
model.add(Dropout(0.2))
model.add(LSTM(units= 50 , return_sequences = False))
model.add(Dense(25))
model.add(Dense(1))
```

```python
keras.utils.plot_model(model = model , to_file = 'AxisBankLSTM.png')
```

```python
model.compile(optimizer='adam' , loss='mean_squared_error')
```

```python
model.fit(train_x,train_y,batch_size=32 , epochs=3)
```

```python
test_data = scaled_data[training_data_len - 60:,:]
test_x = []
test_y = dataset[training_data_len: , :]
for i in range(60,len(test_data)):
    test_x.append(test_data[i-60:i,0])
```

```python
test_x = np.array(test_x)
```
